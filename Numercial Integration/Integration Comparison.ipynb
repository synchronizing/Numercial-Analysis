```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x):
    return (2*x)/(1+x**2)

a = 0
b = 6
N = int(1/.001) # STEP SIZE
n = 10

x = np.linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)

plt.figure(figsize=(40,5))


plt.subplot(1,3,2)
plt.plot(X,Y,'b', label='Actual', linestyle="dashed")

x_mid = (x[:-1] + x[1:])/2
y_mid = f(x_mid)
plt.plot(x_mid,y_mid,'b.',markersize=10, label='Midpoint')
plt.bar(x_mid,y_mid,width=(b-a)/N,alpha=0.2,edgecolor='b')

x = np.linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)

plt.plot(X,Y)

for i in range(N):
    xs = [x[i],x[i],x[i+1],x[i+1]]
    ys = [0,f(x[i]),f(x[i+1]),0]
    plt.fill(xs,ys,'b',edgecolor='b',alpha=0.2)

x = np.linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)

plt.plot(X,Y, label='Trapezoid Rule', linestyle="dashed")

dx = (b-a)/N
x = np.linspace(a,b,N+1)
y = f(x)
S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])

plt.plot(x,y, label='Simpsons Rule', linestyle="dashed")
plt.title('Comparison Graph, N = {}'.format(N))
plt.legend(loc=1)

plt.show()
```

```python
actual = 3.6109179126442244
print("Actual Area: ", actual)

dx = (b-a)/N
x_midpoint = np.linspace(dx/2,b - dx/2,N)
midpoint_riemann_sum = np.sum(f(x_midpoint) * dx)
print("\nMidpoint Riemann Sum:",midpoint_riemann_sum)
print("Difference: ", abs(actual-midpoint_riemann_sum))

y = f(x)
y_right = y[1:] # Right endpoints
y_left = y[:-1] # Left endpoints
dx = (b - a)/N
T = (dx/2) * np.sum(y_right + y_left)
print("\nTrapezoid Area: ", T)
print("Difference: ", abs(actual-T))


print("\nSimpsons Area:", S)
print("Difference: ", abs(actual-S))
```
