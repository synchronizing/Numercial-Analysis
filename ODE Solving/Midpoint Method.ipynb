```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def dy( x, y ):
    return 1/(1+(x**2)) - 2*(y**2)

def y(x):
    return x/(1+x**2)

def midpoint_method( f, a, b, N, IV ):
    h = (b-a)/float(N)
    t = np.arange( a, b+h, h )
    w = np.zeros((N+1,))
    t[0], w[0] = IV
    for i in range(1,N+1):
        w[i] = w[i-1] + h * dy( t[i-1] + h/2.0, w[i-1] + h/2.0 * dy( t[i-1], w[i-1] ))
    return w

a, b = 0.0, 10.0
N = int(1/.001)
h = (b-a)/N
IV = ( 0.0, 0.0 )
t = np.arange( a, b+h, h )
t_full = np.arange(a, b+((b-a)/1000), (b-a)/1000)

 
w = midpoint_method(dy, a, b, N, IV )

midpoint_error = abs(w-y(t)).max()
print("Midpoint's Error: ", midpoint_error)

plt.plot(t, w, label='2nd order')
plt.plot(t_full, y(t_full), label='exact', linestyle="dashed")
plt.title( "Midpoint Method, N="+str(N) )
plt.xlabel('t') 
plt.ylabel('y(t)')
plt.legend(loc=1)
plt.grid()
```
